```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mounting Google Drive
drive.mount('/content/drive')
# Load the dataset from Google Drive
file_path = '/content/drive/MyDrive/govhack/solar_energy_project.csv'
data = pd.read_csv(file_path)

# Univariate analysis of numerical features
numeric_features = ['Lower Nameplate Capacity (MW)', 'Upper Nameplate Capacity (MW)',
                    'Storage Capacity (MWh)', 'latitude', 'longitude']

# Plotting histograms for numerical features
for feature in numeric_features:
    plt.figure(figsize=(10, 5))
    sns.histplot(data[feature], bins=20, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()

# Univariate analysis of categorical features
categorical_features = ['Region', 'Technology Type', 'Fuel Type', 'Unit Status']

# Plotting count plots for categorical features
for feature in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=data, x=feature, order=data[feature].value_counts().index)
    plt.title(f'Count of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Scatter plot of latitude vs. longitude to visualize plant locations
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='longitude', y='latitude', hue='Region', palette='Set2')
plt.title('Geographical Distribution of Solar Power Plants')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Boxplot of Nameplate Capacity by Region
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Region', y='Upper Nameplate Capacity (MW)', palette='Set3')
plt.title('Capacity of Solar Plants by Region')
plt.xlabel('Region')
plt.ylabel('Upper Nameplate Capacity (MW)')
plt.xticks(rotation=45)
plt.show()

# Heatmap to explore correlations between numeric features
plt.figure(figsize=(10, 6))
sns.heatmap(data[numeric_features].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Pairplot for multivariate analysis of key numerical and categorical features
sns.pairplot(data, vars=['Upper Nameplate Capacity (MW)', 'latitude', 'longitude'],
             hue='Region', palette='Spectral')
plt.suptitle('Multivariate Analysis of Capacity, Latitude, and Longitude', y=1.02)
plt.show()

# Exploring the relationship between capacity, region, and technology type
plt.figure(figsize=(14, 8))
sns.scatterplot(data=data, x='longitude', y='latitude', size='Upper Nameplate Capacity (MW)',
                hue='Technology Type', sizes=(20, 200), alpha=0.6, palette='Dark2')
plt.title('Impact of Technology Type on Capacity and Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(loc='upper right')
plt.show()
```
